# TESTING PROCESS

## Importing Libraries

```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

## Labels dictionary

```python
class_names = {0:'Acne',
               1:'benign', 
               2:'Dermatitis',
               3:'Eczema'}
```

## Model Loading

```python
model=load_model("model/CNN_model.h5")
```

## User Input Image

```python
IMG_SIZE = 256
```

```python
img_path="UserInput/benign1.jpg"
```

```python
img=cv.imread(img_path,1)        
img = np.array(cv.resize(img, (IMG_SIZE,IMG_SIZE)))
img = img.astype('float32')
img /= 255.0
```

```python
plt.imshow(img)
plt.show()
```

## Normalization

```python
print(np.shape(img))
```

```python
expand_img = np.expand_dims(img, axis=0)
```

```python
print(np.shape(expand_img))
```

```python
results = model.predict(expand_img)
```

```python
print(results)
```

```python
single_result=results[0]
```

```python
print(single_result)
```

```python
most_likely_class_index = int(np.argmax(single_result))
print(most_likely_class_index)
```

```python
class_label = class_names[most_likely_class_index]
print(class_label)
```

## Final Result

```python
print("Predicted class is -------> {} ".format(class_label))
```
